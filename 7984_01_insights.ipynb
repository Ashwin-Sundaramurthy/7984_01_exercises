```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("movie_metadata.csv")
imdb= pd.read_csv("imdb.csv")
```

```python
data_content_rating=data['content_rating'].value_counts(dropna=False)
data_content_rating.plot(kind="bar")
```

```python
data['profit']=data['gross']-data['budget']
dat=data.groupby("title_year").agg({'budget':[min,max,'mean']})
#budget allocated for movies
```

```python
#budget over the years
data.groupby("title_year")['budget'].sum().plot()

#increasing over the years
```

```python
#code to get number of movies with search genre string
def genre(data,x) :
    return(sum(data.genres.str.contains(x)))
genre(data,"Comedy")
```
